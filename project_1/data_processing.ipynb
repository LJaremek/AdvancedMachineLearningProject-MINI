# Task 1 - Data processing

```python
def plot_dataset(dataset, title: str = None) -> None:
    plt.subplots(figsize=(20, 10))
    sns.heatmap(dataset.corr(), cmap="seismic", center=0).set_title(title)
    title = title.lower().replace(" ", "_")
    plt.savefig(f"plots/processing/{title}.png")
    plt.show()
```

```python
def check_correlation(data):
    correlation_matrix = data.corr().abs()
    strongly_correlated = (
        (correlation_matrix > 0.8) & (correlation_matrix < 1.0)
    )

    correlated_columns = []
    for col in strongly_correlated.columns:
        correlated_cols = strongly_correlated.index[strongly_correlated[col]]
        correlated_cols = correlated_cols.tolist()
        if correlated_cols:
            correlated_columns.append((col, correlated_cols))

    return correlated_columns
```

## Large datasets

### 1. Ionosphere - https://archive.ics.uci.edu/ml/datasets/ionosphere

```python
ionosphere_data = pd.read_csv(
    "data/ionosphere/ionosphere.data",
    header=None,
    index_col=None
    )

ionosphere_data.describe()
```

```python
print("Number of null values:")
print(ionosphere_data.isnull().sum().sum()) 

correlated_columns = check_correlation(ionosphere_data)

for col, correlated_cols in correlated_columns:
    msg = f"Column {col} is strongly correlated with columns: "
    msg += f"{', '.join(map(str, correlated_cols))}"
    print(msg)
            
plot_dataset(ionosphere_data, title="Ionsphere dataset")
```

```python
ionosphere_data = ionosphere_data.drop(14, axis=1)
```
